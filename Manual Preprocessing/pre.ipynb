# Replace missing values
 

1.categorical -> replace by mode (most frequently occuring value)
2.numerical -> replace by mean or median


```python
X["AirBags"].value_counts()
```

```python
mode = X["AirBags"].mode()[0]
mode
```

```python
X["AirBags"].isna().sum()
```

```python
# fillna function to replace missing values
X["AirBags"] = X["AirBags"].fillna(mode)
```

```python
X["AirBags"].isna().sum()
```

```python
X.head()
```

```python
X["Luggage.room"].mean()
```

```python
X["Luggage.room"].median()
```

```python
def replacer(df: pd.DataFrame):
    cat_cols = df.select_dtypes(include="object").columns.tolist()

    for col in df.columns:
        if col in cat_cols:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)

    print("Missing values replaced in dataframe")
```

```python
X.isna().sum()
```

 # Step 5 - Apply Feature Engineering


Catgorical features -> OneHotEncoding


```python
X_cat = X[cat_cols]
X_num = X[num_cols]
```

```python
X_cat
```

```python
X_num
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False, drop="first"
).set_output(transform="pandas")

X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre
```

# 2. Apply StandardScaler


```python
X_num
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")
X_num_pre = scaler.fit_transform(X_num)
X_num_pre.head()
```

# Combine categorical and numerical columns


```python
X_pre = X_num_pre.join(X_cat_pre)
X_pre.head()
```

# Data preprocessing complete


# Step 5 - Train Test Split


80 % - Train data, 20% as Test Data

Train data used for model training
Test data used for model evaluation

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.2, random_state=21
)
```

# Step 6 - Model building

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

# Step 7 - Evaluate the modelStep 7 - Evaluate the model

```python
# R2 score in train
r2_train = model.score(xtrain, ytrain)
print(r2_train)
```

```python
# R2 score in test
r2_test = model.score(xtest, ytest)
print(r2_test)
```

```python
# generalization error
gen_err = abs(r2_train - r2_test)
print(gen_err)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
)


def evaluate_model(model, x, y):
    # Predict the results
    ypred = model.predict(x)
    # Get the metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print the results
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print(f"Training Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print(f"Testing Results : ")
evaluate_model(model, xtest, ytest)
```
